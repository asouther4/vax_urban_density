To begin, we import all the packages we need. We use pandas for data manipulation, matplotlib for visualization, and econtools for econometrics work. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import econtools.metrics as mt
import csv
```

We will also define a few helpful functions to make repetitive data tasks a bit cleaner. 

```python
#this function imports an experiment data log file and performs some simple cleaning
def import_and_clean(filepath):
    
    df = pd.read_json("experiment_data_logs/homophily_test_10_26.log", lines=True)
    
    df["proportion_vacc"] = 1 - (df["unvaccinated"] / df["hub_size"])
    df["proportion_inf"] = (df["recovered"] / df["hub_size"]) 
    df["hub_id"] = df["hub"].astype(str) + df["inst_unique_id"] #generate a unique id for each hub

    return df
```

### Experiment 1: The Effect of Homophily

First, import the data and generate two useful variables.

```python
df1 = import_and_clean("experiment_data_logs/homophily_test_10_26.log")
print(df1.shape)
df1.head()
```

Split the data into three smaller dataframes based on the homophily parameter. Then split each of these dataframes further into "urban" and "rural" data.

```python
#create a new df only where homophily is 99%
df_99 = df1.query("homophily==.99")
df_99.reset_index
#split into urban and rural data
df_urban_99 = df_99.loc[df_99["hub_density"]==12 ]
df_rural_99 = df_99.loc[df_99["hub_density"]==8]

#create another df only where homophily is 89%
df_89 = df1.query("homophily==.89")
df_89.reset_index
#split into urban and rural data
df_urban_89 = df_89.loc[df_89["hub_density"]==12 ]
df_rural_89 = df_89.loc[df_89["hub_density"]==8]

#create another df only where homophily is 79%
df_79 = df1.query("homophily==.79")
df_79.reset_index
#split into urban and rural data
df_urban_79 = df_79.loc[df_79["hub_density"]==12 ]
df_rural_79 = df_79.loc[df_79["hub_density"]==8]

#look at the shape of each dataframe
print(df_99.shape)
print(df_89.shape)
print(df_79.shape)
```

```python
df_99 = df1.query("homophily==.99")
df_89 = df1.query("homophily==.89")
df_79 = df1.query("homophily==.79")
print(df_99.shape)
print(df_89.shape)
print(df_79.shape)
```

Before any data analysis, we want to visualize these epidemics. For this experiment, we need six new dataframes to build some line plots. We divide the data into six subsets based on homophily and hub contact rate. Then, for each epidemic season, we average all the data over every hub. 

```python
urban_99_grouped = df1.query("homophily==.99 and hub_density==12").groupby(["season"]).mean()
rural_99_grouped = df1.query("homophily==.99 and hub_density==8").groupby(["season"]).mean()

urban_89_grouped = df1.query("homophily==.89 and hub_density==12").groupby(["season"]).mean()
rural_89_grouped = df1.query("homophily==.89 and hub_density==8").groupby(["season"]).mean()

urban_79_grouped = df1.query("homophily==.79 and hub_density==12").groupby(["season"]).mean()
rural_79_grouped = df1.query("homophily==.79 and hub_density==8").groupby(["season"]).mean()

#let's take a look at one of the dataframes for good measure
urban_99_grouped.head()
```

Build three line plots to compare urban vs. rural vaccination rates across the three degrees of homophily.

```python
#build the basic structure of our line plots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (18,4), sharey=True)
x = urban_99_grouped.index

#first line plot
axes[0].plot(x, urban_99_grouped["proportion_vacc"], color='#7f6d5f', label='Urban')
axes[0].plot(x, rural_99_grouped["proportion_vacc"], color='#557f2d', label='Rural')
axes[0].set_title("Avg Regional Vaccination over Time, Homophily = 0.99")
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Vaccination Rate")
axes[0].legend(loc="best")

#second line plot
axes[1].plot(x, urban_89_grouped["proportion_vacc"], color='#7f6d5f', label='Urban')
axes[1].plot(x, rural_89_grouped["proportion_vacc"], color='#557f2d', label='Rural')
axes[1].set_title("Avg Regional Vaccination over Time, Homophily = 0.89")
axes[1].set_xlabel("Season")
axes[1].legend(loc="best")

#third line plot
axes[2].plot(x, urban_79_grouped["proportion_vacc"], color='#7f6d5f', label='Urban')
axes[2].plot(x, rural_79_grouped["proportion_vacc"], color='#557f2d', label='Rural')
axes[2].set_title("Avg Regional Vaccination over Time, Homophily = 0.79")
axes[2].set_xlabel("Season")
axes[2].legend(loc="best")
```

Build similar graphs for infection rates

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (18,4), sharey=True)

axes[0].plot(x, urban_99_grouped["proportion_inf"], color='#7f6d5f', label='Urban')
axes[0].plot(x, rural_99_grouped["proportion_inf"], color='#557f2d', label='Rural')
axes[0].set_title("Avg Regional Infection over Time, Homophily = 0.99")
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Infection Rate")
axes[0].legend(loc="best")

axes[1].plot(x, urban_89_grouped["proportion_inf"], color='#7f6d5f', label='Urban')
axes[1].plot(x, rural_89_grouped["proportion_inf"], color='#557f2d', label='Rural')
axes[1].set_title("Avg Regional Infection over Time, Homophily = 0.89")
axes[1].set_xlabel("Season")
axes[1].legend(loc="best")

axes[2].plot(x, urban_79_grouped["proportion_inf"], color='#7f6d5f', label='Urban')
axes[2].plot(x, rural_79_grouped["proportion_inf"], color='#557f2d', label='Rural')
axes[2].set_title("Avg Regional Infection over Time, Homophily = 0.79")
axes[2].set_xlabel("Season")
axes[2].legend(loc="best")
```

#### Experiment 1: Summary Statistics

```python
#only data from the last five seasons
exp1_last5 = df.loc[df["season"]>=20]

#generate a unique id for each hub
exp1_last5["hub_id"] = exp1_last5["hub"].astype(str) + exp1_last5["inst_unique_id"]

#collapse the data to the hub level, averaging over the last 5 seasons
exp1_hubs = exp1_last5[["proportion_inf", 
                          "proportion_vacc", 
                          "homophily", 
                          "hub_density",
                          "hub_id"]].groupby(["hub_id"]).mean().round(2)

#clean up some of the variable names for the table
exp1_hubs["Percent Vaccinated"] = exp1_hubs["proportion_vacc"]*100
exp1_hubs["Percent Infected"] = exp1_hubs["proportion_inf"]*100
exp1_hubs = exp1_hubs.rename(columns={"homophily": "Homophily",
                                    "hub_density": "Hub Contact Rate"})

#build table 1
table1 = exp1_hubs.groupby(["Homophily", "Hub Contact Rate"]).agg({"Percent Infected": ['mean', 'std'],
                                                                  "Percent Vaccinated": ['mean', 'std']}).round(2)
#export table1 to csv and display it
table1.to_csv("tables\homophily_exp_table.csv")
table1
```

#### Difference in Differences Estimation

```python
exp1_dum = pd.get_dummies(exp1_hubs[["Hub Contact Rate", "Homophily"]], columns=["Hub Contact Rate", "Homophily"])
exp1_dum = exp1_dum.rename(columns={"Hub Contact Rate_8": "Rural",
                                "Hub Contact Rate_12": "Urban"})

reg1 = pd.concat([exp1_dum, exp1_hubs], axis=1) 
df_79_99 = reg1.loc[reg1["Homophily"]!=0.89]
df_79_99["Homophily_0.99:Urban"] = df_79_99["Homophily_0.99"]*df_79_99["Urban"]


x = ["Homophily_0.99", "Urban", "Homophily_0.99:Urban"]
y = "Percent Vaccinated"
results = mt.reg(df_79_99, y, x)

print(results)
```

### Experiment 2: The Effect of Density

```python
df2 = pd.read_json("experiment_data_logs/density_test_10_27.log", lines=True)
df2["proportion_vacc"] = 1 - (df2["unvaccinated"] / df2["hub_size"])
df2["proportion_inf"] = (df2["recovered"] / df2["hub_size"])
print(df2.shape)
df2.head()
```

```python
#find the datapoints where density is 6 and 14
df_urban_14 = df2.loc[df2["hub_density"]==14 ]
df_rural_6 = df2.loc[df2["hub_density"]==6]


#find the next set of datapoints where density is 8 and 12
df_urban_12 = df2.loc[df2["hub_density"]==12 ]
df_rural_8 = df2.loc[df2["hub_density"]==8]


#last, find the data where density is 10
df_10 = df2.loc[df2["hub_density"]==10 ]

#average out the data over each season
df_14_grouped = df_urban_14.groupby(["season"]).mean()
df_6_grouped = df_rural_6.groupby(["season"]).mean()
df_12_grouped = df_urban_12.groupby(["season"]).mean()
df_8_grouped = df_rural_8.groupby(["season"]).mean()
df_10_grouped = df_10.groupby(["season"]).mean()
```

```python
#generate all the needed x values
r1 = df_14_grouped.index

fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (18,4), sharey=True)
axes[0].plot(r1, df_14_grouped["proportion_vacc"], color='#7f6d5f', label='Urban: 14 contacts')
axes[0].plot(r2, df_6_grouped["proportion_vacc"], color='#557f2d', label='Rural: 6 contacts')
axes[0].set_title("Regional Vaccination over Time: 14 vs 6 Contact Rate")
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Average Vaccination Rate")
axes[0].legend(loc="best")

axes[1].plot(r1, df_12_grouped["proportion_vacc"], color='#7f6d5f', label='Urban: 12 contact')
axes[1].plot(r2, df_8_grouped["proportion_vacc"], color='#557f2d', label='Rural: 8 contacts')
axes[1].set_title("Regional Vaccination over Time: 12 vs 8 Contact Rate")
axes[1].set_xlabel("Season")
axes[1].legend(loc="best")

axes[2].plot(r1, df_10_grouped["proportion_vacc"], color='#7f6d5f', label='10 contacts')
axes[2].set_title("Regional Vaccination over Time: 10 Contact Rate")
axes[2].set_xlabel("Season")
axes[1].legend(loc="best")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (18,4), sharey=True)
axes[0].plot(r1, df_14_grouped["proportion_inf"], color='#7f6d5f', label='Urban: 14 contacts')
axes[0].plot(r2, df_6_grouped["proportion_inf"], color='#557f2d', label='Rural: 6 contacts')
axes[0].set_title("Regional Infection over Time: 14 vs 6 Contact Rate")
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Average Infection Rate")
axes[0].legend(loc="best")

axes[1].plot(r1, df_12_grouped["proportion_inf"], color='#7f6d5f', label='Urban: 12 contact')
axes[1].plot(r2, df_8_grouped["proportion_inf"], color='#557f2d', label='Rural: 8 contacts')
axes[1].set_title("Regional Infection over Time: 12 vs 8 Contact Rate")
axes[1].set_xlabel("Season")
axes[1].legend(loc="best")

axes[2].plot(r1, df_10_grouped["proportion_inf"], color='#7f6d5f', label='10 contacts')
axes[2].set_title("Regional Infection over Time: 10 Contact Rate")
axes[2].set_xlabel("Season")
axes[1].legend(loc="best")
```

```python
#only data from the last five seasons
exp2_last5 = df2.loc[df2["season"]>=20]

#generate a unique id for each hub
exp2_last5["hub_id"] = exp2_last5["hub"].astype(str) + exp2_last5["inst_unique_id"]

#collapse the data to the hub level, averaging over the last 5 seasons
exp2_hubs = exp2_last5[["proportion_inf", 
                          "proportion_vacc", 
                          "homophily", 
                          "hub_density",
                          "hub_id"]].groupby(["hub_id"]).mean().round(2)

#clean up some of the variable names for the table
exp2_hubs["Percent Vaccinated"] = exp2_hubs["proportion_vacc"]*100
exp2_hubs["Percent Infected"] = exp2_hubs["proportion_inf"]*100
exp2_hubs = exp2_hubs.rename(columns={"homophily": "Homophily",
                                    "hub_density": "Hub Contact Rate"})

#create a variable for each treatment
exp2_hubs.loc[(exp2_hubs["Hub Contact Rate"]==6) | (exp2_hubs["Hub Contact Rate"]==14), "Treatment"] = "A"
exp2_hubs.loc[(exp2_hubs["Hub Contact Rate"]==8) | (exp2_hubs["Hub Contact Rate"]==12), "Treatment"] = "B"
exp2_hubs.loc[(exp2_hubs["Hub Contact Rate"]==10), "Treatment"] = "C"


#build table 2
table2 = exp2_hubs.groupby(["Treatment", "Hub Contact Rate"]).agg({"Percent Infected": ['mean', 'std'],
                                                                  "Percent Vaccinated": ['mean', 'std','count']}).round(2)
#export table2 to csv and display it
table2.to_csv("tables\contact_rate_exp_table.csv")
table2
```

```python
#only data from the last five seasons
exp4a_last5 = df4a.loc[df4a["season"]>=20]
exp4b_last5 = df4b.loc[df4b["season"]>=20]

#generate a unique id for each hub
exp4a_last5["hub_id"] = exp4a_last5["hub"].astype(str) + exp4a_last5["inst_unique_id"]
exp4b_last5["hub_id"] = exp4b_last5["hub"].astype(str) + exp4b_last5["inst_unique_id"]

#collapse the data to the hub level, averaging over the last 5 seasons
exp4a_hubs = exp4a_last5[["proportion_inf", 
                          "proportion_vacc", 
                          "hub_density",
                          "hub_id"]].groupby(["hub_id"]).mean().round(2)

exp4b_hubs = exp4b_last5[["proportion_inf", 
                          "proportion_vacc", 
                          "hub_density",
                          "hub_id"]].groupby(["hub_id"]).mean().round(2)


#clean up some of the variable names for the table
exp4a_hubs["Percent Vaccinated"] = exp4a_hubs["proportion_vacc"]*100
exp4a_hubs["Percent Infected"] = exp4a_hubs["proportion_inf"]*100
exp4a_hubs = exp4a_hubs.rename(columns={"homophily": "Homophily",
                                    "hub_density": "Hub Contact Rate"})

exp4b_hubs["Percent Vaccinated"] = exp4b_hubs["proportion_vacc"]*100
exp4b_hubs["Percent Infected"] = exp4b_hubs["proportion_inf"]*100
exp4b_hubs = exp4b_hubs.rename(columns={"homophily": "Homophily",
                                    "hub_density": "Hub Contact Rate"})


#create a variable for each treatment
exp4a_hubs.loc[(exp4a_hubs["Hub Contact Rate"]==6) | (exp4a_hubs["Hub Contact Rate"]==14), "CR Treatment"] = "A"
exp4a_hubs.loc[(exp4a_hubs["Hub Contact Rate"]==8) | (exp4a_hubs["Hub Contact Rate"]==12), "CR Treatment"] = "B"
exp4a_hubs.loc[(exp4a_hubs["Hub Contact Rate"]==10), "CR Treatment"] = "C"

exp4b_hubs.loc[(exp4b_hubs["Hub Contact Rate"]==6) | (exp4b_hubs["Hub Contact Rate"]==14), "CR Treatment"] = "A"
exp4b_hubs.loc[(exp4b_hubs["Hub Contact Rate"]==8) | (exp4b_hubs["Hub Contact Rate"]==12), "CR Treatment"] = "B"
exp4b_hubs.loc[(exp4b_hubs["Hub Contact Rate"]==10), "CR Treatment"] = "C"


#build table 4a
vacc_2 = exp2_hubs.groupby(["Treatment", "Hub Contact Rate"]).agg({"Percent Vaccinated": ['mean', 'std']}).round(2)
vacc_2 = vacc_2.rename(columns={"Percent Vaccinated": "Percent Vaccinated (Pop Size I)"})
vacc_4a = exp4a_hubs.groupby(["CR Treatment", "Hub Contact Rate"]).agg({"Percent Vaccinated": ['mean', 'std']}).round(2)
vacc_4a = vacc_4a.rename(columns={"Percent Vaccinated": "Percent Vaccinated (Pop Size II)"})
vacc_4b = exp4b_hubs.groupby(["CR Treatment", "Hub Contact Rate"]).agg({"Percent Vaccinated": ['mean', 'std']}).round(2)
vacc_4b = vacc_4b.rename(columns={"Percent Vaccinated": "Percent Vaccinated (Pop Size III)"})

#concatenate the tables together
table4a = pd.concat([vacc_2,vacc_4a,vacc_4b], axis=1)

#export table4a to csv and display it
table4a.to_csv("tables\pop_size_vacc_table.csv")
table4a
```

```python
#now we want to create the same table as above for inf rates
#build table 4a
inf_2 = exp2_hubs.groupby(["Treatment", "Hub Contact Rate"]).agg({"Percent Infected": ['mean', 'std']}).round(2)
inf_2 = inf_2.rename(columns={"Percent Infected": "Percent Infected (Pop Size I)"})
inf_4a = exp4a_hubs.groupby(["CR Treatment", "Hub Contact Rate"]).agg({"Percent Infected": ['mean', 'std']}).round(2)
inf_4a = inf_4a.rename(columns={"Percent Infected": "Percent Infected  (Pop Size II)"})
inf_4b = exp4b_hubs.groupby(["CR Treatment", "Hub Contact Rate"]).agg({"Percent Infected": ['mean', 'std']}).round(2)
inf_4b = inf_4b.rename(columns={"Percent Infected": "Percent Infected (Pop Size III)"})

#concatenate the tables together
table4b = pd.concat([inf_2,inf_4a,inf_4b], axis=1)

#export table4a to csv and display it
table4b.to_csv("tables\pop_size_inf_table.csv")
table4b
```
